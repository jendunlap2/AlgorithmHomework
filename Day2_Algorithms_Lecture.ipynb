## Two Pointers

#### Syntax

```python
# lst[left], lst[right] = lst[right], lst[left]
# continue to swap until left index os greater than the right index (they pass eachother)

def reverse_in_place(lst):
    # Create the pointers for our indices
    left = 0
    right = len(lst) - 1
    while left <= right:
        lst[left], lst[right] = lst[right], lst[left]
        left += 1
        right -= 1
    
some_list = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]

print(some_list)
reverse_in_place(some_list)
print(some_list)
```

# Sorting Algorithms

#### Bubble Sort

Worst Case: O(n^2) Time - O(1) Space

```python
def bubble_sort(array):
    is_sorted = False
    while not is_sorted:
        is_sorted = True
        for i in range (len(array)-1):
            if array[i] > array[i+1]:
                array[i], array[i+1] = array[i+1], array[i]
                is_sorted = False
    return array


```

```python
import random
random_list = [random.randint(0,100) for i in range(10)]
print(random_list)
```

```python
bubble_sort(random_list)
```

##### Insertion Sort

Worst Case: O(n^2) time - O(1)space

```python
def insertion_sort(array):
    for i in range(1, len(array)):
        while i > 0 and array[i] < array[i-1]:
            array[i], array[i-1] = array[i-1], array[i]
            i-=1


random_list = [random.randint(0,100) for i in range(10)]
print(random_list)

insertion_sort(random_list)
print(random_list)
```

## Merge Sort

#### How it Works

```python
# Step 1: Split everything into its own group down to len == 1 

# Step 2: From left to right, conitinue to merge two groups together

# Step 3: While merging, we will plaace each item in the correct index position in the main list

# Step 4: Continue until we have only 1 group left


def merge_sort(alist):
    if len(alist) > 1:
        mid = len(alist) // 2
        print("Splitting...:", alist)
        left_half = alist[:mid]
        right_half = alist[mid:]
        
        merge_sort(left_half)
        merge_sort(right_half)
        
        # index pointers fror our list
        l = 0 # pointer for left_half list
        r = 0 # pointer for right_half list
        m = 0 # pointer for main list
        
        while l < len(left_half) and r < len(right_half):
            if left_half[l] < right_half[r]:
                alist[m] = left_half[l]
                l+=1
            else:
                alist[m] = right_half[r]
                r+=1
            m+=1
        while l < len(left_half):
            alist[m] = left_half[l]
            l += 1
            m += 1
        while r < len(right_half):
            alist[m] = right_half[r]
            r += 1
            m += 1
        
    print(alist)
    return alist

random_list = [68, 61, 82, 81, 76, 28, 59, 7, 90, 1]
merge_sort(random_list)
```

# Binary Search

The Binary Search algorithm works by finding the number in the middle of a given array and comparing it to the target. Given that the array is sorted

* The worst case run time for this algorithm is `O(log(n))`

```python
def binary_search(alist, target):
    high = len(alist) - 1
    low = 0
    while low <= high:
        middle = (low + high) //2
        if target == alist[middle]:
            return f"The index for target is {middle}."
        elif target < alist[middle]:
            high = middle - 1
        else:
            low = middle + 1
    return -1

random_list = [1, 7, 28, 59, 61, 68, 76, 81, 82, 90]
binary_search(random_list, 82)
```

<p>Create a function that counts how many distinct words are in the string below, then outputs a dictionary with the words as the key and the value as the amount of times that word appears in the string.<br>Should output:<br>{'a': 5,<br>

```python
a_text = 'In computing, a hash table hash map is a data structure which implements an associative array abstract data type, a structure that can map keys to values. A hash table uses a hash function to compute an index into an array of buckets or slots from which the desired value can be found'

# don't forget to start with the Empty Dictionary
distinct_words = {}

# change string to all to lowercase and alphabetize
a_text = sorted(a_text.lower().split())

# Remove punctuation assign as key values to dictionary; Count occurances, assign count to corresponding key value
for i in a_text:
    i = i.strip(',''.')
    distinct_words[i] = a_text.count(i)

# Print the new list
for k, v in distinct_words.items():
    print(f"{k}: {v}")
```
